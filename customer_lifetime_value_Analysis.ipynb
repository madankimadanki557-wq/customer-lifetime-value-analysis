```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500

df = pd.DataFrame({
    "Customer_ID": range(1001, 1001 + n),
    "Purchase_Frequency": np.random.randint(1, 20, n),
    "Average_Order_Value": np.random.randint(50, 1000, n),
    "Customer_Age": np.random.randint(18, 65, n),
    "Years_as_Customer": np.random.randint(1, 10, n)
})

df["CLV"] = (
    df["Purchase_Frequency"]
    * df["Average_Order_Value"]
    * df["Years_as_Customer"]
)

df.head()
```

```python
df.to_csv("customer_clv_data.csv", index=False)
```

```python
df["CLV"].mean()
```

```python
df.groupby("Customer_Age")["CLV"].mean()
```

```python
import matplotlib.pyplot as plt

top10 = df.sort_values("CLV", ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10["Customer_ID"].astype(str), top10["CLV"])
plt.title("Top 10 Customers by CLV")
plt.xlabel("Customer ID")
plt.ylabel("CLV")
plt.show()
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500

df = pd.DataFrame({
    "Customer_ID": range(1001, 1001 + n),
    "Purchase_Frequency": np.random.randint(1, 20, n),
    "Average_Order_Value": np.random.randint(50, 1000, n),
    "Customer_Age": np.random.randint(18, 65, n),
    "Years_as_Customer": np.random.randint(1, 10, n)
})

df["CLV"] = (
    df["Purchase_Frequency"]
    * df["Average_Order_Value"]
    * df["Years_as_Customer"]
)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["CLV"].mean()
```

```python
df.loc[df["CLV"].idxmax()]
```

```python
top10 = df.sort_values("CLV", ascending=False).head(10)

top10
```

```python
import matplotlib.pyplot as plt

top10 = df.sort_values("CLV", ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10["Customer_ID"].astype(str), top10["CLV"])

plt.title("Top 10 Customers by CLV")
plt.xlabel("Customer ID")
plt.ylabel("CLV")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.hist(df["CLV"], bins=20)

plt.title("Customer Lifetime Value Distribution")
plt.xlabel("CLV")
plt.ylabel("Number of Customers")

plt.show()
```
